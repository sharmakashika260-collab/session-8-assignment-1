```python
steps = [2000, 4000, 8000, 10000, 12000, 14000, 16000]
days = ["Sun", "Mon", "Tues", "Wed", "Thurs", "Fri", "Sat"]

i = 0

while i < len(steps):
    if steps[i] > 10000:
        print("The first day when you crossed 10,000 steps is", days[i])
        break
    i = i + 1
```

```python
item_prices = [500, 800, 1000, 1500, 2000]

final_total = 0

for price in item_prices:
    if price == 0:
        continue

    final_total = final_total + price

    if final_total > 2000:
        break

print(final_total)
```
